# Data Exploration - Pre Processed Dataset

```python
import pandas as pd

original_data = pd.read_csv('../data/original.csv')
```

### Missing values

```python
pd.options.mode.use_inf_as_na = True

def print_missing_values(data):
    print('Missing values: \n')
    for i in data:
        count = 0
        for j in data[i]:
            if j == 'None' or j == None or j == '' or j == ' ':
                count += 1
        if (count != 0):
            print(f'{i}: {count} ; {round(count / len(data) * 100, 2)}%')
```

##### Original dataset

```python
print_missing_values(original_data)
```

### Graph Review Lenght

```python
import matplotlib.pyplot as plt
import numpy as np

lenghts = []
x = [x for x in range(0,len(original_data))]

for i in original_data['Reviews']:
    lenghts.append(len(i.split()))

plt.axhline(y=np.nanmean(lenghts), color='red', linestyle='--', linewidth=3, label='Avg')
plt.scatter(x, lenghts,3)

plt.title('Number of words per review')    
plt.xlabel('Review Index')
plt.ylabel('Number of words')

plt.show()

print(f'Average: {np.nanmean(lenghts)}')
```
